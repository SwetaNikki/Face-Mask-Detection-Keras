```python
import os, cv2
import numpy as np
from tensorflow.keras.preprocessing import image
for i in os.listdir(test_path):
    img = image.load_img(test_path+"//"+i, target_size=(100,100))
    plt.imshow(img)
    plt.show()
    
    test_data = []
    img=cv2.imread(test_path)

    try:
        gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)           
            #Coverting the image into gray scale
        resized=cv2.resize(gray,(100,100))
            #resizing the gray scale into 100x100, since we need a fixed 
        test_data.append(resized)
        #target.append(label_dict[category])

    except Exception as e:
            print()
            #if any exception rasied, print nothing. And pass to the next image
    
    test_data=np.array(test_data)/255.0
    test_data=np.reshape(test_data,(test_data.shape[0],100,100,1))
    images = np.vstack([test_data])
    result = model.predict(images)
    

    
    if result == 1:
        print("Without Mask")
        print()
        print()
    else:
        print("With Mask")
        print()
        print()
```
